# CLEAN FILE ORGANIZER 

```python
# importing necessary files 
import os,shutil,pathlib
```

```python
# variables 
root1 = 'c:\\users\\homepc'
root2 = 'g:\\'

DO,DOC = 'Downloads','Documents'
PER,WOR = 'Personal','Work'
AU,CO = 'Audio','Compressed'
VD,PIC = 'Videos','Pictures'
OR,CP = 'Original','Copy'
DOF,PDF = 'Docx files','Pdf files'
EF,S = 'Excel files','Slides'
O,N = 'Other','Notes'
```

```python
# File Extensions 
FILE_EXTENSIONS = {  
	
               "Pictures": [".jpeg", ".jpg", ".tiff", ".gif",
                           ".bmp", ".png", ".bpg", "svg",".heif", ".psd"], 
	
               "Videos": [".avi", ".mkv",".flv", ".wmv",
                         ".mov", ".mp4", ".webm", ".vob", 
                         ".mng",".qt", ".mpg", ".mpeg", ".3gp"], 
	
              "Documents": [".oxps", ".epub", ".pages", ".docx",
                           ".txt", ".pdf", ".doc", ".fdf",
                           ".ods",".odt", ".pwi", ".xsn",
                           ".xps", ".dotx", ".docm", ".dox",
                           ".rvg", ".rtf", ".rtfd", ".wpd", 
                           ".xls", ".xlsx", ".ppt","pptx"], 
	
               "Compressed": [".a", ".ar", ".cpio", ".iso", 
                              ".tar", ".gz", ".rz", ".7z",
                              ".dmg", ".rar", ".xar", ".zip"], 
	
               "Audio": [".aac", ".aa", ".aac", ".dvf",
                         ".m4a", ".m4b", ".m4p", ".mp3",
                         ".msv", "ogg", "oga", ".raw", 
                         ".vox", ".wav", ".wma"],
                         }
```

```python
# Creating paths
DOWN = os.path.join( root1,DO )
DOWN_PER = os.path.join( root1,DO,PER )
DOWN_WOR = os.path.join( root1,DO,WOR )

DOC_PER = os.path.join( root2,PER )
DOC_WOR = os.path.join( root2,WOR )
```

```python
# FUNCTIONS

 # Personal file transfers
def Personal_transfer( folder , source , destination , extension ):

    if folder != 'Documents':

        x = os.path.join( destination , folder )  
        shutil.move( source , x )             # Moving it to the target folder 

    else:

        format_to_folder = {

            '.docx' : DOF ,
            '.docm' : DOF ,
            '.xls' : EF ,
            '.xlsx' : EF ,
            '.pdf' : S ,
            '.txt' : N
        }
        
        target_folder = format_to_folder.get( extension , O )     # getting the target folder 
        
        x = os.path.join( destination , DOC , target_folder )       #getting the target destination
        
        shutil.move( source , x )      # Moving to the Target Destination 


#function that moves , copies and renames work files 
def move_and_rename( source , original_destination  , original_subfolder , copy_destination , copy_subfolder , full_file_name , file_name , extension ):
 
    x = os.path.join( original_destination , original_subfolder )      #original destination
    y = os.path.join( copy_destination , copy_subfolder )   #copy destination

    shutil.copy( source , y )  #copying the file 

    #renaming the file 
    old_file_name = os.path.join( y , full_file_name )
    new_file_name = os.path.join( y , f'{file_name}_copy{extension}')
    os.rename( old_file_name , new_file_name )

    shutil.move( source , x )       #moving the file to originals folder

  
#work file transfers 
def Work_transfer( folder , source , destination , full_file_name , file_name , extension ):

    if folder != 'Documents':

        original_dest = os.path.join( destination , OR )   #destination for original files
        copy_dest = os.path.join( destination , CP )       #destination to copy the files

        move_and_rename( source, original_dest , folder , copy_dest , folder , full_file_name , file_name , extension )
 

    else:

        format_to_folder = {

            '.docx' : DOF ,
            '.docm' : DOF ,
            '.xls' : EF ,
            '.xlsx' : EF ,
            '.pdf' : PDF ,
            '.txt' : N,
            '.ppt': S, 
            '.pptx': S
        }
        
        original_dest = os.path.join( destination , OR , DOC )
        copy_dest = os.path.join( destination , CP , DOC )

        target_folder = format_to_folder.get( extension , O )   #matching the extension with the folder
        
        move_and_rename( source , original_dest , target_folder , copy_dest , target_folder , full_file_name , file_name , extension )
```

```python
#Scanning the Downloads Directory
for folder in os.scandir( DOWN ):

    if folder.is_dir() and folder.name == 'Personal':

        source_dir = DOWN_PER
        
        for file in os.scandir( source_dir ):

            file_source_path = pathlib.Path(file)
            file_format = file_source_path.suffix.lower()
            full_file_name = file_source_path.name
            file_name = file_source_path.stem

            for folder_name,formats in FILE_EXTENSIONS.items():

                if file_format in formats:

                    dest_dir = DOC_PER

                    Personal_transfer( folder_name , file_source_path , dest_dir , file_format )

    elif folder.is_dir() and folder.name == 'Work':
        
        source_dir = DOWN_WOR

        for file in os.scandir( source_dir ):

            file_source_path = pathlib.Path(file)
            file_format = file_source_path.suffix.lower()
            full_file_name = file_source_path.name
            file_name = file_source_path.stem

            for folder_name,formats in FILE_EXTENSIONS.items():

                if file_format in formats:

                    dest_dir = DOC_WOR

                    Work_transfer( folder_name , file_source_path , dest_dir , full_file_name , file_name , file_format )
```
